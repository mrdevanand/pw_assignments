```python
<!DOCTYPE html>
<html>
<head>
    <title>HTML Tags List</title>
    <style>
        body {
            font-family: Arial, sans-serif;
            padding: 30px;
            background-color: #f9f9f9;
        }

        h1 {
            color: #333;
        }

        ol {
            background-color: #fff;
            padding: 20px;
            border-radius: 8px;
            box-shadow: 0 0 5px rgba(0,0,0,0.1);
        }

        li {
            margin-bottom: 15px;
            line-height: 1.5;
        }

        code {
            background-color: #eef;
            padding: 2px 5px;
            border-radius: 4px;
            font-family: monospace;
        }
    </style>
</head>
<body>

    <h1>üìÑ Ordered List of HTML Tags</h1>

    <ol>
        <li><code>&lt;html&gt;</code> ‚Äì The root element of an HTML document.</li>
        <li><code>&lt;head&gt;</code> ‚Äì Contains metadata about the document, like title and links to styles.</li>
        <li><code>&lt;title&gt;</code> ‚Äì Sets the title of the document shown in the browser tab.</li>
        <li><code>&lt;body&gt;</code> ‚Äì Contains all the visible content of the web page.</li>
        <li><code>&lt;h1&gt; to &lt;h6&gt;</code> ‚Äì Header tags for defining headings (h1 is largest, h6 is smallest).</li>
        <li><code>&lt;p&gt;</code> ‚Äì Paragraph tag used to write blocks of text.</li>
        <li><code>&lt;a&gt;</code> ‚Äì Anchor tag used to create hyperlinks.</li>
        <li><code>&lt;img&gt;</code> ‚Äì Embeds images into the page using the `src` attribute.</li>
        <li><code>&lt;ul&gt;</code> ‚Äì Creates an unordered (bulleted) list.</li>
        <li><code>&lt;ol&gt;</code> ‚Äì Creates an ordered (numbered) list.</li>
        <li><code>&lt;li&gt;</code> ‚Äì List item tag used inside <code>&lt;ul&gt;</code> or <code>&lt;ol&gt;</code>.</li>
        <li><code>&lt;div&gt;</code> ‚Äì Generic container for grouping content.</li>
        <li><code>&lt;span&gt;</code> ‚Äì Inline container for styling small pieces of text.</li>
        <li><code>&lt;br&gt;</code> ‚Äì Inserts a line break.</li>
        <li><code>&lt;input&gt;</code> ‚Äì Creates form input fields like textboxes, checkboxes, etc.</li>
        <li><code>&lt;form&gt;</code> ‚Äì Container for form elements to collect user input.</li>
        <li><code>&lt;button&gt;</code> ‚Äì Represents a clickable button.</li>
        <li><code>&lt;table&gt;</code> ‚Äì Defines a table.</li>
        <li><code>&lt;tr&gt;</code> ‚Äì Table row inside a table.</li>
        <li><code>&lt;td&gt;</code> ‚Äì Table data cell inside a row.</li>
    </ol>

</body>
</html>
```
